- Write a class function ``play_game`` that accepts two index values and plays the game. If the character values in selected cells are equal thencards will overturned. 
- Write a class function ``display`` that shows the current state of the game. Initially all cells are showing just numerical values. Look at sample run to understand the initial state of the game. 
- Define a class function that checks the status of the game, if all the cards are overturn will announce the winner. 


Then write a main module that creates an instance of ``Memory_game`` object to play this game.

````python
import random
import copy 
````

````python
def display(lst):
    k=0
    for i in range(4):
        for j in range(4):
            print lst[k], #print(lst[k], end=" ")
            k+=1
        print '\n'
````

````python
def takeInput():
    n1 = input("Enter one number between 0 and 15: ")
    if (n1<0 and n1>15):
        return takeInput()
    
    n2 = input("Enter one number between 0 and 15: ")
    if (n2<0 and n2>15):
        return takeInput()
    
    if n1==n2:
        return takeInput()
    
    else:
        return n1, n2
````

````python
def play_game(disp_lst, memory_lst, actual_lst):

    k=0
    while k<10:
        k+=1
        print "Round ", k
        print "\n"
        n1, n2 = takeInput()

        if actual_lst[n1] == actual_lst[n2]:
            disp_lst[n1] = actual_lst[n1]
            disp_lst[n2] = actual_lst[n1]
            display(disp_lst)
            
        else:
            memory_lst[n1] = actual_lst[n1]
            memory_lst[n2] = actual_lst[n2]
            display(memory_lst)
            print "\n"
            display(disp_lst)
            
        memory_lst = copy.deepcopy(disp_lst)    
        print "\n\n"
        
    if actual_lst == disp_lst:
        print "Congrats, You won the game!!!"
    else:
        print "Too bad, You lost the game!!!"
````

````python
actual_lst = list('AABBCCDDEEFFGGHH')
random.shuffle(actual_lst)

disp_lst = []
memory_lst = []

for i in range(16):
    disp_lst.append(i)
    memory_lst.append(i)
    
play_game(disp_lst, memory_lst, actual_lst)
````
